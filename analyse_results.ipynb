```python
import json
import os
import pandas as pd
```

```python
output_dir = "./output"
data = {}
for fname in os.listdir(output_dir):
    fpath = os.path.join(output_dir, fname)
    if fpath.endswith(".json"):
        data_dict = json.load(open(fpath, "r"))
        data[fname] = {k: (v["answer"], round(v["score"], 3)) for k, v in data_dict.items()}
```

```python
df = pd.DataFrame(data).T
```
